# Combining Dataset
Now that the individual dataset is sufficiently cleaned we can proceed to merge the 3 different dataset into a signle datase 

```python
data = pd.merge(sales,store, left_on = 'Store',right_on='Store')
```

```python
data.shape
```

```python
data.info()
```

```python
data = pd.merge(data,feature, left_on = ['Date','Store'],right_on=['Date','Store'])
```

```python
np.sum([data.IsHoliday_x == data.IsHoliday_y])
#As the 2 columns are identical we can remove one of them
```

```python
data = data.drop('IsHoliday_y',axis =1)
```

```python
data.info()
```

```python
#Data is clean. Exporting data to a csv for future usage
data.to_csv('./Data/combined data.csv')
```

```python
#Converting date to datetime object
data.Date = pd.to_datetime(data.Date, dayfirst=True)
```
